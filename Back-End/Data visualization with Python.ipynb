```python
import pandas as pd
import numpy as np
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_json('nyTimesData_2021_1.json')
```

```python
type(data)
```

```python
articles = pd.DataFrame(data.response.docs)
```

```python
type(articles)
```

```python
articles
```

```python
allHeadlines = ''
for h in articles.headline:
    allHeadlines = allHeadlines + ' ' + h['main']
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width=1600, height=800,stopwords=stopwords,max_words=50, background_color="white").generate(allHeadlines)
wordcloud.to_file("nyTimesJanuary2021_wordcloud.png")
```

```python
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
